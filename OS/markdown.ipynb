# 운영체제

## 01. 오리엔테이션

### 1.1. 운영체제 거시적으로 보기

1. kernel: 프로세스 관리, 메모리 관리, 저장공간 관리, 장치 관리 등 컴퓨터에 속한 자원들에 대한 접근을 중재하는 역할을 함. 

2. interface: 사용자의 명령을 컴퓨터에 전달하고 결과를 사용자에게 알려주는 소통의 역할. 사용자가 직접 접하는 화면이 여기에 속함. 운영체제가 제공하는 대표적인 인터페이스는 GUI(Graphical User Interface), CLI(Command Line Interface)가 있음.

3. system call: 사용자나 프로그램이 직접적으로 컴퓨터 자원에 접근하는 것을 막고 커널을 보호하기 위해서 만든 코드 집합. system call 함수를 통해 커널에 접근.

4. driver: 프린터, 키보드 및 디스크 드라이브와 같은 하드웨어 장치와 운영체제 간의 통신을 가능하게 하는 소프트웨어.

- 리소스 : 실행에 마땅히 필요한 요소
    - CPU, Memory, File , etc.

- 운영체제 : 리소스들을 움직이게 하는 것, 자원을 관리하고 할당하는 특별한 프로그램

    - CPU관리, 프로세스 관리, 파일 시스템 관리, 메모리 관리 .. 

<p align="center"><img src="image/01.png" width="800"></p>


- 응용프로그램은 하드웨어에 바로 접근이 불가능하고 운영체제를 통해 접근이 가능하다.

    - 시스템 콜 : 운영체제 서비스를 제공받기 위해 커널 모드로 전환하는 것

    - 소프트웨어 인터럽트

- 이중 모드

    - 커널 모드 : 운영체제 서비스를 제공받을 수 있는 모드 (입출력 가능)

    - 사용자 모드 : 운영체제 서비스를 제공받을 수 없는 모드 (입출력 불가능)

- 운영체제 서비스

    - 프로세스 관리

    - 자원 관리 접근 및 할당(cpu, memory, 보조기억장치, 입출력장치)

    - 파일 시스템 관리


### 1.3. strace 기반 시스템 콜 관찰하기
- 리눅스 환경 > mac은 dtruss 왜 안되는지 모르겟음

- close : 파일을 닫는 시스템 콜(파일 관련 자원 해제)

- getpid/getppid : PID를 반환하는 시스템 콜

- syslog : 시스템 로크 메세지 남기기

- exit : 실행 중인 프로그램 종료

## 02. 프로세스와 스레드

### 2.1. 커널 영역과 사용자 영역의 프로세스

#### 프로세스

- 실행중인 프로그램 > 같은 프로그램도 별도의 프로세스가 될 수 있다.

- 포그라운드 프로세스

- 백그라운드 프로세스
    
    - 서비스, 데몬 : 사용자와 상호작용하지 않으면서 실행되는 프로세스

- 프로세스 제어 블록(PCB)

    -  운영 체제(OS)가 프로세스를 관리하는 데 사용하는 자료구조다. 프로세스의 핵심 정보를 포함하며, OS가 프로세스를 생성, 스케줄링, 종료하는 데 필수적인 역할을 한다. PCB는 프로세스의 상태, 메모리, 파일, 입출력 장치 등 다양한 정보를 보관한다.

    - PID(Process ID): 프로세스를 식별하는 고유 번호

    - 레지스터: CPU가 현재 실행 중인 명령어와 관련된 정보를 저장

    - 스케줄링 정보: 프로세스의 우선 순위, 실행 상태, CPU 사용 시간 등 스케줄링에 필요한 정보를 포함

    - 메모리 정보: 프로세스가 사용하는 메모리 영역, 가상 메모리 주소 공간 등 메모리 관련 정보를 보관

    - 사용한 파일 정보: 프로세스가 열어둔 파일 목록, 파일 접근 권한 등 파일 관련 정보를 담음

    - 입출력 장치 정보: 프로세스가 사용하는 입출력 장치, 장치 드라이버 정보 등 입출력 관련 정보를 포함

- 문맥교환

    - 문맥(context) : 실행을 재개하기 위해 기억해야 할 정보

    - 문맥 교환 : 여러 프로세스들이 번갈아가며 실행되는 원리(백업, 복구)

## 03. CPU 스케줄링

### 3.1. 프로세스 우선순위와 스케줄링 큐

- 모든 프로세스(및 스레드)는 실행되기 위해 자원을 필요로 한다.

- 운영체제가 공정하고 합리적으로 자원을 배분하는 방법 == 스케줄링

- 모든 프로세스(및 스레드)는 실행되기 위해 CPI를 필요로 한다.

- 운영체제가 공정하고 합리적으로 CPU를 배분하는 방법 == CPU 스케줄링

- CPU 자원은 한정되어 있고 실행중인 프로세스가 여러 개 일 때 프로세스들이 CPU를 나누는 방법
    - 프로세스마다 우선순위(PCB에 명시)가 다르다

    - $ps -el > PRI, NI : 낮을수록 높은 우선순위

    - $TOP > PR, NI : 낮을수록 높은 우선순위

    - I/O bound process > CPU bound process
        - I/O bound process : 입출력 장치 사용하는 시간이 비교적 많음
            - 조금만 실행하면 입출력 작업이 완료될때까지 대기상태에 진입하기 때문에 빠르게 처리하므로 우선순위가 높다
        - CPU bound process : CPU를 사용하는 시간이 비교적 많음 
            - 컴파일 작업

- 프로세스 우선순위를 토대로 CPU 할당 받는 방법 == CPU 스케줄링 알고리즘

#### 선점형 스케줄링과 비선점형 스케줄링

- 한 프로세스 실행 도중 다른 급한 프로세스가 실행되어야 한다면
    - 선점형 스케줄링
        - 현재 실행중인 프로세스의 자원을 빼앗아 해당 프로세스에 할당

        - 프로세스에 자원을 고루 할당 가능

        - 문맥 교환 과정의 오버헤드

        - 실시간 시스템, 사용자 인터페이스, 멀티미디어 애플리케이션

    - 비선점형 스케줄링
        - 현재 실행 중인 프로세스 실행이 끝날 때까지 해당 프로세스 대기(비선점형)

        - 고르지 않은 자원 분배

        - 문맥 교환 과정에서의 오버헤드 적음

        - 프린터 드라이버, 웹 서버, 배치 처리 시스템


- 응답성 요구 사항: 응답성이 중요한 경우 선점형 스케줄링을 사용해야 합니다.

- 공정성 요구 사항: 모든 작업이 공정하게 처리되어야 하는 경우 비선점형 스케줄링을 사용해야 합니다.

- 시스템 부하: 시스템 부하가 높은 경우 선점형 스케줄링이 더 효율적일 수 있습니다.

- 프로세스 특성: 프로세스의 특성에 따라 선점형 스케줄링 또는 비선점형 스케줄링이 더 적합할 수 있습니다.


### 3.2. CPU 스케줄링 알고리즘
* 전공서 기준 CPU 스케줄링 알고리즘

| 알고리즘 | 장점 | 단점 | 적합한 상황 |
|---|---|---|---|
| 선입 선처리 (FIFO) | - 구현이 간단 | - 호위 효과 발생 | - 단순한 배치 시스템 |
| 최단 작업 우선 (SJF) | - 평균 대기 시간 최소화 | - 예측 불가능, 부정확한 예상 실행 시간 | - 짧고 예측 가능한 작업 |
| 라운드 로빈 (RR) | - 공정성, 빠른 응답 | - CPU 집중형 프로세스 불리 | - 대화형 시스템, 웹 서버 |
| 최소 잔여 시간 우선 (SRT) | - SJF + RR 장점 | - 예상 실행 시간 필요 | - SJF와 RR 장점이 필요한 경우 |
| 우선순위 | - 중요한 작업 우선 처리 | - 아사 현상 발생 | - 중요한 작업 우선 처리 필요 |
| 다단계 큐 | - 다양한 프로세스 유형 처리 | - 구현 복잡 | - 다양한 프로세스 유형 처리 |
| 다단계 피드백 큐 | - 다단계 큐 + 아사 현상 해결 | - 구현 복잡 | - 다양한 프로세스 유형, 아사 현상 방지 필요 |


1. 선입 선처리 스케줄링(FIFO 스케줄링)

    - CPU를 먼저 요청한 프로세스부터 CPU 할당

    - 준비 큐에 삽인된 순서대로 실행되는 비선점형 스케줄링

    - 부작용 : 호위 효과 (convoy effect)

2. 최단 작업 우선 스케줄링 (SJF 스케줄링)

    - 준비 큐 프로세스 중 CPU 이용 시간이 짧은 프로세스부터 실행

    - 호위 효과 방지

3. 라운드 로빈 스케줄링

    - 선입 선처리 스케줄링 + 타임 슬라이스

    - 준비 큐에 삽입된 순서로 실행하되, 타임 슬라이스만큼 실행

    - 선점형 스케줄링

4. 최소 잔여 시간 우선 스케줄링(SRT)

    - 최단 작업 우선 스케줄링 + 라운드 로빈 스케줄링

    - 작업 시간 짧은 프로세스부터 처리하되, 타임 스라이스만큼 돌아가면 실행

5. 우선순위 스케줄링 (범용적)

    - 프로세스마다 우선순위 부여, 우선순위 높은 순으로 스케줄링

    - 아사 (starvation) 현상
        - 모든 우선순위 스케줄링 알고리즘의 근본적인 문제

        - 우선순위 낮은 프로세스의 실행이 계속 연기되는 현상

        - 우선순위 높은 프로세스를 실행하느라 우선순위가 낮은 프로세스를 실행하지 못한다

        - 해결책 : 에이징(aging) > 대기 시간이 길어지면 점차 우선순위를 높이는 방식

6. 다단계 큐 스케줄링

    - 우선순위별로 준비 큐를 여러 개 사용하는 스케줄링

    - 프로세스 유형 별로 큐 구분 가능
        e.g. cpu 바운드, I/O 바운드, 백그라운드, 포그라운드, 실시간 프로세스 ..
    
    - 큐 별로 다른 스케줄링 알고리즘 적용 가능
        e.g. 선입 선처리 큐, 라운드 로빈 큐, ..
    
    - 큐 별로 다른 타임 슬라이스 적용 가능

    - 기본적으로 프로세스는 큐간의 이동이 불가능하여 아사 현상 발생 가능성이 있다

7. 다단계 피드백 큐 스케줄링

    - 다단계 큐 스케줄링의 아사 현상을 해결하기 위한 스케줄링

    - 프로세스가 큐 간의 이동 가능

    - CPU bound 프로세스 자연스럽게 우선순위가 내려감

    - I/O bound 프로세스는 높은 우선순위에서 끝낼 수 있다

    - 높은 우선순위 큐에 삽입, 실행이 끝나지 않을 경우 낮은 우선순위 큐에 삽입

    - 에이징 적용

    - CPU bound 프로세스, I/O bound 프로세스 구분 가능

### 4.3. 뮤텍스와 세마포어

#### **동기화 해결을 위한 세 가지 원칙**

1. 상호 배제 : 한프로세스가 임계 구역에 진입했다면 다른 프로세스는 대기해야 함

2. 진행 : 어떤 프로세스도 임계 구역에 진입하지 않았다면 집입이 가능해야 함

3. 유한 대기 : 한 프로세스가 임계 구역 진입을 위해 대기하고 있다면 언젠가 진입이 가능해야 함

#### 뮤택스 락
- 상호 배제를 위한 동기화 도구

- 한 마디로 자물쇠(lock)

- 자물쇠 역할 : 프로세스들이 공유하는 전역변서 lock

- 자물쇠 잠그기 : acquire 함수

- 자물쇠 열기 : release 함수

- 임계구역 전에 호출하는 함수

- 스레드가 실행될 때 acquire 함수로 락 > 잠겨 있을시 while문으로 계속 확인(바쁜대기) > 다른 스레드가 임계구역에서 작업을 마쳤다면 lock은 false로 바꾸는 release 함수 실행

#### 세마포(semaphore)
- 상호 베체 & 실행 순서 제어를 위한 동기화 도구

##### 카운팅 세마포

- 뮤텍스 락은 기본적으로 공유 자원이 하나일 경우 상징

- 세마포는 공유 자원이 여러 개 있을 경우도 동기화 가능 * 실행 순서 제어, 상호 배제 동기화

- 가라는 신호를 받으면 임계 구역에 진입해도 좋다

- 멈추라는 신호면 임계 구역에 진입하면 안된다

1. 변수 S : 임계 구역에 진입할 수 있는 프로세스의 개수(사용 가능한 공유 자원의 개수)
    

2. wait 함수 : 임계 구역에 들어가도 좋은지, 기다려야 하는지 알려주는 함수
    - 만일 임계구역에 진입할 수 있는 프로세스의 개수가 0이하라면

    - 사용할 수 있는 자원이 있는지 반복적으로 확인

    - 임계구역에 진입할 수 있는 프로세스가 1개 이상이라면 S를 1감소시키고 임계구역 진입

3. signal 함수 : 임계 구역 앞에서 기다리는 프로세스에게 이제 가도 좋다는 신호를 주는 함수
    
    - 임계구역에서의 작업을 마친 뒤 S 1 증가

### 4.6. 교착상태와 해결 방법

- A 프로세스가 x자원을 사용하고있고 y자원을 기다리고있다, B 프로세스는 y자원을 사용하고 있고 x자원을 기다리고있을 때 교착상태가 일어난다

<img src="image/22.png" alt="800">

- 철학자가 원탁에 둘러 앉아있고, 각 철학자들은 생각하는 행동, 밥 먹는 행동 두가지 행동을 할 수 있다.

- 배가 고파지면 철학자들은 자신의 왼쪽, 오른쪽의 젓가락을 모두 집어서 밥을 먹는다.

- 정해진 시간동안 식사를 한다

- 양쪽 젓가락을 내려놓는다

- 처음부터 다시 반복한다

- 철학자 == 프로세스, 스레드

- 포크 == 실행을 위해 필요한 자원

- 발생 가능한 문제 : 교착상태 deadlock
    - 아무것도 진행되지 못하고 막혀있는 상황이 생길 수 있는 문제
    e.g. 모든 철학자가 모두 배가고파서 왼쪽 젓가락을 잡아버리면 아무도 오른쪽 젓가락을 잡을 수 없어서 아무도 밥을 먹을 수 없다.


##### 교착 상태 발생할 수 있는 조건

- 상호 배제 : 동시에 자원을 사용할 수 없는 상태

- 점유와 대기 : 자원을 할당 받은 채 다른 자원의 할당을 기다리는 상태

- 비선점 : 강제로 자원을 빼앗을 수 없는 경우

- 원형 대기 : 자원을 원형으로 대기할 경우

#### 해결 방법

- 예방

- 회피

- 검출 후 회복

##### 교착상태 예방

- 교착 상태 방생 조건 네 가지 중 하나를 없애는 것

- 교착 상태가 발생 배경 원천 차단

- 교착 상태가 발생하지 않음을 보장할 수 있지만, 여러 부작용이 따르는 방식

1. 상호 배제 조건 없애기

- 자원을 공유 가능하도록 변경

- 모든 자원에 대해 적용할 수 없음(e.g.프린터기)

2. 점유와 대기 조건 없애기

- 특정 프로세스에 자원을 모두 할당, 아예 할당하지 않기

- 자원의 활용률 저하(효율적인 분배가 안된다)

3. 비선점 조건 없애기

- 선점하여 사용 가능한 자원에 대해서는 효과적 (e.g.CPU)

- 모든 자원에 대해 적용 가능한 것은 아님 (e.g.프린터기)

4. 원형 대기 조건 없애기

- 자원에 번호 매기기

- 오름차순으로 자원 할당
